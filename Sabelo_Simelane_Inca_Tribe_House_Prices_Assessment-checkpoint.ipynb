## Data Preprocessing

###### Checking for missing values

```python
df.isna().sum()
```

###### Imputing the missing values with their column means

```python
df['Area']=df['Area'].fillna(df['Area'].mean())
df['Bedrooms']=df['Bedrooms'].fillna(df['Bedrooms'].mean())
df['Bathrooms']=df['Bathrooms'].fillna(df['Bathrooms'].mean())
```

###### Change the price column to float, so that it matches the datatypes of the other numeric columns

```python
df['Price'] = df['Price'].astype(float)
```

```python
df.info()
```

```python
df.isna().sum()
```

###### Check categorical columns for the percentage of their data that is labelle unknown

```python
columns = ['Furnished', 'Level', 'Compound', 'Payment_Option', 'Delivery_Date', 'Delivery_Term', 'City']

for column in columns:
    unknown_count = len(df[df[column] == 'Unknown'])
    unknown_ratio = unknown_count / len(df)
    print(column + ':', unknown_ratio)
```

###### Remove columns with more than 35% of their data labelled as unknown

```python
columns_to_drop = ['Level', 'Compound', 'Delivery_Date']
df = df.drop(columns_to_drop, axis=1)
df.head()
```

## Model Building

```python
column_order = ['Type', 'Bathrooms', 'Area', 'Bedrooms', 'Furnished', 'Payment_Option', 'Delivery_Term', 'City', 'Price']
df = df[column_order]
```

```python
X = df.iloc[:, 0:8].values
y = df.iloc[:, 8].values
```

## Encoding

```python
label_encoder_Type = LabelEncoder()
label_encoder_Furnished = LabelEncoder()
label_encoder_Payment_Option = LabelEncoder()
label_encoder_Delivery_Term = LabelEncoder()
label_encoder_City = LabelEncoder()
```

```python
X[:,0] = label_encoder_Type.fit_transform(X[:,0])
X[:,4] = label_encoder_Furnished.fit_transform(X[:,3])
X[:,5] = label_encoder_Payment_Option.fit_transform(X[:,4])
X[:,6] = label_encoder_Delivery_Term.fit_transform(X[:,5])
X[:,7] = label_encoder_City.fit_transform(X[:,6])
```

## Scaling

```python
Scaler = StandardScaler()
X_standard = Scaler.fit_transform(X)
y_standard = Scaler.fit_transform(y.reshape(-1,1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_standard, y_standard, test_size = 0.3, random_state = 0)
```
